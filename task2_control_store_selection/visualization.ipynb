```python
import pandas as pd
import matplotlib.pyplot as plt

chips_joined = pd.read_csv("chips_joined.csv")

# Aggregate weekly sales
weekly_sales = chips_joined.groupby("WEEK")["TOT_SALES"].sum().reset_index()

# Plot
plt.figure(figsize=(12,6))
plt.plot(weekly_sales["WEEK"], weekly_sales["TOT_SALES"], marker="o")
plt.title("Weekly Sales Trend")
plt.xlabel("Week")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load segment metrics
segment_metrics = pd.read_csv("segment_metrics.csv")

# Plot
plt.figure(figsize=(12,6))
sns.barplot(
    data=segment_metrics,
    x="LIFESTAGE",
    y="total_sales",
    hue="PREMIUM_CUSTOMER",
    estimator=sum
)

plt.title("Revenue by Customer Segment")
plt.ylabel("Total Sales ($)")
plt.xlabel("Lifestage")
plt.xticks(rotation=30, ha="right")
plt.legend(title="Premium Customer")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load pack size metrics
pack_size_metrics = pd.read_csv("pack_size_metrics.csv")

# Sort by pack size for a clean x-axis
pack_size_metrics = pack_size_metrics.sort_values("PACK_SIZE")

# Plot
plt.figure(figsize=(12,6))
sns.barplot(
    data=pack_size_metrics,
    x="PACK_SIZE",
    y="total_sales",
    color="skyblue"
)

plt.title("Sales by Pack Size")
plt.ylabel("Total Sales ($)")
plt.xlabel("Pack Size (g)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load promo impact data
promo_impact = pd.read_csv("promo_impact.csv")

# Plot
plt.figure(figsize=(6,6))
sns.barplot(
    data=promo_impact,
    x="PROMO_FLAG",
    y="total_sales",
    palette="pastel"
)

plt.title("Promo vs Non-Promo Sales")
plt.ylabel("Total Sales ($)")
plt.xlabel("Promo Flag")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load brand sales data
brand_sales = pd.read_csv("brand_sales.csv")

# Sort by sales (already sorted, but just to be safe)
brand_sales = brand_sales.sort_values("total_sales", ascending=False)

# Plot
fig, ax1 = plt.subplots(figsize=(12,6))

# Bar chart for sales
ax1.bar(brand_sales["BRAND"], brand_sales["total_sales"], color="skyblue")
ax1.set_xlabel("Brand")
ax1.set_ylabel("Total Sales ($)", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")
ax1.set_xticklabels(brand_sales["BRAND"], rotation=45, ha="right")

# Line chart for cumulative %
ax2 = ax1.twinx()
ax2.plot(brand_sales["BRAND"], brand_sales["cum_sales_pct"], color="red", marker="o", linestyle="-")
ax2.set_ylabel("Cumulative Sales %", color="red")
ax2.tick_params(axis="y", labelcolor="red")
ax2.axhline(80, color="gray", linestyle="--")  # 80% reference line

plt.title("Pareto Analysis of Brand Sales")
plt.tight_layout()
plt.show()
```

 Executive Summary (Draft)
- Overall Sales: Chips generated ~$1.77M in sales across ~246k transactions.
- Top Segments: Midage Singles/Couples  Mainstream and Older Families  Budget/Mainstream are the strongest contributors.
- Pack Sizes: 175g packs dominate sales, followed by 150g and 170g. Smaller packs (70g, 90g) and extralarge packs (330g, 380g) are niche.
- Promotions: No promo data was available in the dataset, so promo impact analysis could not be conducted.
- Brands: Sales are highly concentrated  Kettle, Smiths, Doritos, and Pringles together account for over 50% of sales. The top ~9 brands contribute ~80% (Pareto effect).
